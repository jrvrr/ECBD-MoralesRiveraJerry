```python
import pandas as pd
```

```python
import random
```

```python
#crear un data frame con 1,000,000 de registros
cantidad = 1000000
```

```python
influencer = ["luisito", "vegetta", "sof", "emmaPR", "cheinbaum", "montoya", "el wey del pan", "Diego"]

# likes = random.randint(1000,10000000)

comentarios = ["wow" , "🌴🌴🌴" ,"que bonito" ,"gpi" ,"gpi" ,"mi bello cancun" ,"estan funando a jd" ,"bot bot", "es  ia", "hater #1" , "primer comentario", "<3"]

# fecha = random.randint(2024,2026)

tipo_post = ["foto", "video", "tag", "share"]
```

```python
# creacion del DataFrame

dfMillon = pd.DataFrame({
	"influencer" : [ random.choice(influencer) for _ in range(cantidad)],
	"likes" : [ random.randint(1000,10000000) for _ in range(cantidad)],
	"comentarios" : [ random.choice(comentarios) for _ in range(cantidad)],
	"fecha" : [ random.randint(2024,2026) for _ in range(cantidad)],
	"tipo_post" : [ random.choice(tipo_post) for _ in range(cantidad)]
})
```
