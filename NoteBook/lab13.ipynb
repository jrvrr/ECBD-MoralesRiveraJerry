```python
#Importamos las librerias
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.datasets import load_iris
```

```python
#Cargamos el dataset iris
iris = load_iris(as_frame = True)
```

```python
print (iris.DESCR)
```

```python
dfIris = iris.frame
```

```python
dfIris.head()
```

# Exploracion del conjunto de datos

```python
dfIris.shape
```

```python
dfIris.columns
```

```python
iris.target_names
```

```python
##Renombrar la columna "target"
dfIris.rename(columns = {"target": "especie"}, inplace = True)
```

```python
dfIris["especie"] = dfIris["especie"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
```

```python
dfIris.head()
```

```python
dfIris.tail()
```

# Identificar variables predictoras y variables de objetivo

```python
#Variable predictora
X =dfIris.drop(columns=["especie"])
```

```python
print("Variables predictoras")
print(X.columns)
```

```python
#Variable objetivo
y = dfIris["especie"]
```

```python
print("Variable objetivo")
print(y.name)
```

# Dividir los datos en prueba y entrenamiento

80% para entrenamiento

20% para pruebas

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.20,
    random_state = 42, stratify = y
)
```

```python
print ("X_train", X_train.shape)
print ("X_test", X_test.shape)
print ("y_train", y_train.shape)
print ("y_test", y_train.shape)
```

# Escalado de variables

```python
scaler = StandardScaler()
```

```python
# Escalamos las variables
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
print("Sin escalar")
print(X_train.head())
print("Escalado")
print(X_train_scaled[:5])
```

# Entrenamiento del Modelo KNN

En esta prueba utilizaremos k=3

```python
modelo = KNeighborsClassifier(n_neighbors=3)
```

```python
modelo.fit(
    X_train_scaled, y_train
)
```

## Realizamos predicciones con las variables de prueba

```python
prediccion = modelo.predict(X_test_scaled)
```

```python
print(prediccion[:10])
```

```python
#Comparacion entre valores reales y valores predecidos
comparacion = pd.DataFrame({
    "especie real": y_test.values,
    "especie predicha": prediccion
})
```

```python
comparacion.head(30)
```

## Evaluacion del modelo

```python
accuracy = accuracy_score(y_test, prediccion)
print (f"accuracy/exactitud: {accuracy: .4f}")
```

```python
#Matriz de confusion
cm = confusion_matrix(y_test, prediccion)
```

```python
cm_df = pd.DataFrame(
    cm,
    index = modelo.classes_,
    columns= modelo.classes_,
    
)
cm_df
```

## Clasification report
proporciona metricas que permiten evaluar el desempeño de cada clase: precision, recall, f1-score, support

```python
print (classification_report(y_test, prediccion))
```
