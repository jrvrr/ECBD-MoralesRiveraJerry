# Contexto del Problema

Una institución financiera desea analizar las características de distintos individuos para determinar qué factores están asociados con ingresos superiores a $50,000 dólares anuales.

El objetivo será construir un modelo capaz de predecir si una persona pertenece a uno de los siguientes grupos:

- (<=) 50K
- (>) 50K

5. ¿Por qué fue necesario transformar variables categóricas?
Porque los algoritmos de Machine Learning (como el DecisionTreeClassifier que usamos) operan mediante cálculos matemáticos y ecuaciones, por lo que no pueden interpretar texto directamente. Transformarlas a números les permite procesar la información.

7. ¿Cuántas columnas adicionales se generaron después del One-Hot Encoding?
El dataset original tiene variables categóricas con muchas opciones. Al aplicar pd.get_dummies(x, drop_first=True), cada categoría única se convierte en una columna nueva binaria (0 o 1). Se generan aproximadamente entre 90 y 100 columnas en total, dependiendo de las categorías únicas presentes en tu archivo.

11. ¿Cuántos registros quedaron en entrenamiento?
El dataset total tiene 32,561 registros. Al separar el 80% para entrenamiento (test_size=0.20), quedan exactamente 26,048 registros para entrenar.

12. ¿Cuántos registros quedaron en prueba?
Al separar el 20% restante para la evaluación, quedan exactamente 6,513 registros en el conjunto de prueba.

13. ¿Por qué no debemos entrenar utilizando todos los datos?
Porque si usamos el 100% de los datos para entrenar, no tendríamos cómo evaluar de forma justa el modelo. Estaríamos evaluándolo con los mismos datos que ya vio, provocando un problema llamado sobreajuste (overfitting) (el modelo memoriza las respuestas en lugar de aprender a generalizar). Separar datos de prueba nos permite simular cómo actuará el modelo con datos nuevos que nunca ha visto en el mundo real.

14. ¿Cuál fue el Accuracy obtenido?
Al correr el código con un árbol de decisión limitado (max_depth=10), el Accuracy obtenido suele rondar entre el 84% y 85% (0.845 - 0.855).

15. ¿Consideras que el resultado es adecuado?
Sí, es adecuado y bastante bueno. Basándonos en la tabla de interpretación que armamos, un resultado entre el 80% y 90% se clasifica como "Muy bueno". Supera por mucho a un modelo que decida al azar (que tendría un 50% de éxito).

16. ¿Qué factores podrían afectar el desempeño del modelo?
Desequilibrio de clases: Hay muchas más personas que ganan <=50K que personas que ganan >50K, lo que hace que el modelo tienda a aprender mejor a identificar a los de bajos ingresos.

Valores atípicos (Outliers): Columnas como capital-gain tienen valores extremos que pueden confundir al árbol.

Profundidad del árbol: Si no se limita el árbol (max_depth), se expandirá demasiado memorizando datos ruidosos.
