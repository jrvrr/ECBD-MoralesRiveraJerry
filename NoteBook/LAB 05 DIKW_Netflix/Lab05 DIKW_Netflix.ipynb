```python
# Importa pandas para cargar, limpiar y manipular la tabla de datos
import pandas as pd
```

```python
# Importa matplotlib para ajustar detalles del gráfico 
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Cargar el dataset original
df = pd.read_csv("netflix_titles.csv")
```

```python
# Mostrar estadísticas descriptivas incluyendo variables de texto
df.describe(include="all")
```

```python
# Limpieza: nombres de columnas en minúsculas y sin espacios a los lados
df.columns = df.columns.str.strip().str.lower()
```

```python
# Verificar que los nombres de las columnas cambiaron correctamente
df.columns
```

```python
# Detectar valores nulos por columna antes de limpiar
df.isnull().sum()
```

```python
# Calcular el porcentaje de valores faltantes por columna
(df.isnull().sum() / len(df)) * 100
```

```python
# Rellenamos director con un valor genérico por exceso de nulos
df["director"] = df["director"].fillna("Desconocido")
```

```python
# Rellenamos country con la palabra No especificado
df["country"] = df["country"].fillna("No especificado")
```

```python
# Rellenamos cast con un texto genérico de información vacía
df["cast"] = df["cast"].fillna("Sin informacion")
```

```python
# Rellenamos el rating con el valor más repetido (moda)
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
# Eliminar las filas donde la columna duration sea nula
df.dropna(subset=["duration"], inplace=True)
```

```python
# Comprobar que ya no existan valores nulos en el dataset
df.isnull().sum()
```

```python
# Detectar registros completamente duplicados en la tabla
df.duplicated().sum()
```

```python
# Convertir texto de fecha a datetime de python ignorando errores
df["date_added"] = pd.to_datetime(df["date_added"].str.strip(), errors="coerce")
```

```python
# Crear columna nueva con el año rellenando vacíos con 0
df["year_added"] = df["date_added"].dt.year.fillna(0).astype(int)
```

```python
# Capa de Información: Gráfica 1 de distribución de películas vs series
plt.figure(figsize=(6, 4))
sns.countplot(x="type", data=df)
plt.title("Gráfica 1: Distribución de Películas vs Series")
plt.xlabel("Tipo de Contenido")
plt.ylabel("Cantidad de Títulos")
plt.show()
```

```python
# Capa de Información: Gráfica 2 de los 10 países con más producciones
plt.figure(figsize=(9, 4))
top_paises = df["country"].value_counts().head(10)
sns.barplot(x=top_paises.values, y=top_paises.index)
plt.title("Gráfica 2: Top 10 Países con Mayor Volumen de Contenido")
plt.xlabel("Total de Títulos")
plt.ylabel("País")
plt.show()
```

```python
# Capa de Información: Gráfica 3 del top de clasificaciones de edad
plt.figure(figsize=(10, 4))
order_ratings = df["rating"].value_counts().index
sns.countplot(x="rating", data=df, order=order_ratings)
plt.title("Gráfica 3: Títulos por Clasificación de Edad")
plt.xlabel("Clasificación")
plt.ylabel("Cantidad de Títulos")
plt.xticks(rotation=45)
plt.show()
```

```python
# Capa de Información: Gráfica 4 sobre la evolución temporal de adición de títulos
plt.figure(figsize=(9, 4))
df["year_added"].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Gráfica 4: Evolución Histórica de Contenido Agregado por Año")
plt.xlabel("Año de Adición")
plt.ylabel("Cantidad de Títulos")
plt.grid(True)
plt.show()
```
