```python
# Limpieza
# nombres de columnas en minúsculas y sin espacios a los lados
df.columns = df.columns.str.strip().str.lower()
```

```python
df.dtypes
```

```python
#rellenar valores nulos
df["director"] = df["director"].fillna("Desconocido")
```

```python
#valores de country, rellenarlos
df["country"] = df["country"].fillna("No especificado")
```

```python
#valores de rating
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Limpieza de nulos restantes para asegurar que no haya datos vacíos
df['cast'] = df['cast'].fillna('Sin informacion')
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
```

```python
# También eliminamos los 3 nulos de duration para dejar el dataset al 100%
df.dropna(subset=['duration'], inplace=True)
```

```python
# Convertimos a fecha y extraemos el año para las gráficas
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
df['year_added'] = df['date_added'].dt.year
```

```python
df.isnull().sum()
```

```python
# Gráfica 1: Distribución de Tipo de Contenido (Películas vs Series)
#El análisis muestra que las Películas superan ampliamente a las Series, lo que indica que Netflix
# sigue siendo una plataforma mayoritariamente de cine, aunque las series generan mucha fidelidad.
# Esto nos permite conocer qué formato prefiere producir o adquirir la plataforma
plt.figure(figsize=(8, 5))
sns.countplot(x='type', data=df, palette='Reds')
plt.title('Distribución de Contenido: Películas vs Series')
plt.show()
```

```python
# Gráfica 2: Top 10 Países con más contenido
#Estados Unidos e India son los mayores productores, lo que sugiere que el contenido en inglés 
# e hindi domina el catálogo global.
# Nos ayuda a conocer cuáles son los mercados principales de producción
plt.figure(figsize=(10, 5))
df['country'].value_counts().head(10).plot(kind='bar', color='darkred')
plt.title('Gráfica 2: Top 10 Países con mayor contenido')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfica 3: Clasificación de contenido (Ratings)
#La mayoría del contenido tiene clasificación TV-MA (Adultos), lo que concluye que el público 
# objetivo principal de Netflix no son los niños, sino los adultos jóvenes.
# Identificamos a qué tipo de audiencia (edad) se dirige Netflix mayormente
plt.figure(figsize=(10, 5))
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index, palette='viridis')
plt.title('Gráfica 3: Distribución por Clasificación (Rating)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfica 4: Crecimiento histórico de contenido agregado
#Se observa un crecimiento masivo a partir de 2016. Como decisión estratégica, Netflix debería
# invertir más en contenido familiar para competir con otras plataformas como Disney+.
# Analizamos la tendencia de crecimiento de la plataforma por año
plt.figure(figsize=(10, 5))
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='red')
plt.title('Gráfica 4: Títulos agregados por año')
plt.grid(True)
plt.show()
```
