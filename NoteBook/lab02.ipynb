```python
import pandas as pd
```

```python
import random
```

```python
#crear un data frame con 1,000,000 de registros
cantidad = 1000000
```

```python
influencer = ["Luisito Rey", "vegetta777", "EL escorpion Dorado", "Willyrex", "Fernanfloo", "AuronPlay", "Andres manuel", "Flavio"]

# likes = random.randint(1000,10000000)

comentarios = ["Increible" , "El mejorr" ,"que gay" ,"omg" ,"no puede ser","" ,"quien es ese?", "que feo", "vamonos", "mi primer commit" , "¿Ya se termino?", "esto es todo"]

# fecha = random.randint(2020,2026)

tipo_post = ["fotos", "videos", "hashtag", "compartido"]
```

```python
# creacion del DataFrame

dfMillon = pd.DataFrame({
	"influencer" : [ random.choice(influencer) for _ in range(cantidad)],
	"likes" : [ random.randint(1000,10000000) for _ in range(cantidad)],
	"comentarios" : [ random.choice(comentarios) for _ in range(cantidad)],
	"fecha" : [ random.randint(2020,2026) for _ in range(cantidad)],
	"tipo_post" : [ random.choice(tipo_post) for _ in range(cantidad)]
})
```
