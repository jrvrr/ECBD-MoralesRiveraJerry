```python
# libreria de random
import random
```

```python
#crear un data frame con 1,000,000 de registros

cantidad = 1000000
```

```python
#datos aleatorios
nombres = ["Ana", "Luis", "Martha", "Carolina", "Sofia", "Pedro", "Elena", "Diego"]
ciudades = ["CDMX","Monterrey","Guadalajara","Puebla","Cancun","Merida","Tijuana","Toluca"]
categorias = ["Tecnologias" , "ropa" ,"hogar" ,"herramientas" ,"consolas" ,"deportes" ,"calzado" ,"jardineria" ,]
```

```python
# creacion del DataFrame

dfMillon = pd.DataFrame({
	"nombre" : [ random.choice(nombres) for _ in range(cantidad)],
	"edad" : [ random.randint(18,70) for _ in range(cantidad)],
	"salario" : [ round(random.uniform(5000,50000),2) for _ in range(cantidad)],
	"activo" : [ random.choice([True,False]) for _ in range(cantidad)],
	"categorias" : [ random.choice(categorias) for _ in range(cantidad)],
	"ciudades" : [ random.choice(ciudades) for _ in range(cantidad)],
	"anio_registro" : [ random.randint(2020,2026) for _ in range(cantidad)],
})
```

```python
dfMillon.shape
```

```python
#para ver q tipo de dato lo definio
dfMillon.dtypes
```
