```python
# Limpieza
# nombres de columnas en minúsculas y sin espacios a los lados
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
#detectar valores nulos
df.isnull().sum()
```

```python
#eliminar una columna"cabin"
df.drop(columns=["cabin"], inplace=True)
```

```python
df.columns
```

```python

df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos el fare con el valor mas repetido (moda)
df["fare"] = df["fare"].fillna(df["fare"]. mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# graficamos el total de pasajeros por clase
plt.figure(figsize=(8, 4))
sns.countplot(df, x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.ylabel("Total")
plt.show()
```

```python
#graficamos las clases por genero
plt.figure(figsize=(8,4))
sns.countplot(df, x="sex", hue="pclass")
plt.title("Total de genero por clases")
plt.xlabel("Genero")
plt.ylabel("Total")
plt.show()
```

```python
#Mostramos la edad por genero y clase
plt.figure(figsize=(6,4))
sns.violinplot(df,x="sex", y="age", hue="pclass")
plt.title("Rango de edad por genero y clase")
plt.xlabel("Genero")
plt.ylabel("Edad")
plt.show()
```

```python
#Conocimiento
#Los pasajeros de primera clase pagaban tarifas mas altas
df.groupby("pclass")["fare"].mean()
```
